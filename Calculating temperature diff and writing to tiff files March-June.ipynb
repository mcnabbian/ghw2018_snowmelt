```python
import matplotlib.pyplot as plt
import numpy as np
from osgeo import gdal
import rasterio
import rasterio.plot as rioplot
import os
from osgeo import gdal
import shapefile
import shapely.geometry as shp
from rasterio import mask
```

```python
year = np.arange(2014,2017)
month = np.arange(3,7)
var = ["tmean","tmin","tmax"]
nvar = 3
print(month)
print(year)
prism_array = np.zeros([year.shape[0],month.shape[0],621,1405], dtype=np.float)
prism_diff1 = np.zeros([nvar,621,1405], dtype=np.float) 
prism_diff2 = np.zeros([nvar,621,1405], dtype=np.float) 
for vv in np.arange(nvar):
    for yy in np.arange(year.shape[0]):
        for mm in np.arange(month.shape[0]):
           # print(yy)
            #print(mm)
            prism_path = '../data/PRISM_{}_stable_4kmM2_{}0{}_bil.bil'.format(var[vv],year[yy],month[mm])
           # print(prism_path)
            prism_nodata = -9999

            prism_ds = gdal.Open(prism_path)
            prism_band = prism_ds.GetRasterBand(1)
            prism_array[yy,mm,:,:] = prism_band.ReadAsArray().astype(np.float32)
            # prism_nodata = prism_band.GetNoDataValue()
            prism_array[prism_array == prism_nodata] = np.nan
            prism_ds = None
    prism_seasonalmean = np.mean(prism_array,1)
    prism_diff1[vv,:,:] = prism_seasonalmean[1,:,:] - prism_seasonalmean[0,:,:]
    prism_diff2[vv,:,:] = prism_seasonalmean[1,:,:] - prism_seasonalmean[2,:,:]
plt.imshow(prism_diff1[1,:,:], cmap='bwr',vmin = -4, vmax = 4)
plt.title('2015 minus 2014 tmean (March-June)')
plt.colorbar()
plt.show()
```

```python
original_image = rasterio.open('../data/PRISM_tmean_stable_4kmM2_201404_bil.bil')
```

```python
with rasterio.open("../interimfiles/prism_diff_tmean_2015-2014.tif", 'w', driver='GTiff', height=prism_diff1.shape[1],
            width=prism_diff1.shape[2], count=1, dtype=prism_diff1.dtype,
            crs=original_image.crs, transform=original_image.transform) as dest:
    dest.write(prism_diff1[0,:,:],1)
    
with rasterio.open("../interimfiles/prism_diff_tmean_2015-2016.tif", 'w', driver='GTiff', height=prism_diff2.shape[1],
            width=prism_diff2.shape[2], count=1, dtype=prism_diff2.dtype,
            crs=original_image.crs, transform=original_image.transform) as dest:
    dest.write(prism_diff2[0,:,:],1)
```
